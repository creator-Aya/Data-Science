### Subtask:
Create tables showing descriptive statistics for numerical features (e.g., mean, std, min, max) and frequency counts for categorical features in the `X` (features) DataFrame.

```python
print("Descriptive statistics for numerical features:")
print(X[numerical_cols].describe())

print("\nFrequency counts for categorical features:")
for col in categorical_cols:
    print(f"\n--- {col} ---")
    print(X[col].value_counts())
```

# Task
Import plotting libraries (`matplotlib.pyplot` and `seaborn`), then generate histograms for each numerical feature in `X`, bar plots for each categorical feature in `X`, and a bar plot for the target variable `y` to visualize their distributions and identify key characteristics like potential outliers and class imbalances.

## Import Plotting Libraries

### Subtask:
Import necessary libraries for plotting, such as `matplotlib.pyplot` and `seaborn`.


**Reasoning**:
The subtask requires importing the `matplotlib.pyplot` and `seaborn` libraries for plotting. I will use a code block to import them with their conventional aliases, `plt` and `sns`.



```python
import matplotlib.pyplot as plt
import seaborn as sns
print("Matplotlib and Seaborn imported successfully.")
```

**Reasoning**:
The subtask of importing plotting libraries is complete. Now, I will proceed to generate histograms for each numerical feature in the `X` DataFrame, as requested by the main task and the subtask for visualizations.



```python
print("Generating histograms for numerical features...")
num_cols_to_plot = len(numerical_cols)
n_rows = (num_cols_to_plot + 1) // 2 # Arrange in 2 columns

plt.figure(figsize=(15, 5 * n_rows))

for i, col in enumerate(numerical_cols):
    plt.subplot(n_rows, 2, i + 1)
    sns.histplot(X[col].dropna(), kde=True)
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
print("Histograms generated successfully.")
```

**Reasoning**:
With the numerical feature histograms now generated, the next step is to create bar plots for each categorical feature in the `X` DataFrame to visualize their value counts and distributions, as requested by the subtask.



```python
print("Generating bar plots for categorical features...")
num_cat_cols = len(categorical_cols)
n_rows_cat = (num_cat_cols + 1) // 2 # Arrange in 2 columns

plt.figure(figsize=(15, 5 * n_rows_cat))

for i, col in enumerate(categorical_cols):
    plt.subplot(n_rows_cat, 2, i + 1)
    sns.countplot(x=X[col], order=X[col].value_counts().index)
    plt.title(f'Frequency of {col}')
    plt.xlabel(col)
    plt.ylabel('Count')
    plt.xticks(rotation=45, ha='right')

plt.tight_layout()
plt.show()
print("Bar plots for categorical features generated successfully.")
```

**Reasoning**:
Now that the numerical and categorical features have been visualized, the final visualization step is to create a bar plot for the target variable `y` to display its class distribution, completing the visualization subtask.



```python
print("Generating bar plot for target variable...")
plt.figure(figsize=(8, 6))
sns.countplot(x=y.iloc[:, 0], order=y.iloc[:, 0].value_counts().index)
plt.title('Distribution of Target Variable (A16)')
plt.xlabel('Class')
plt.ylabel('Count')
plt.show()
print("Bar plot for target variable generated successfully.")
```
